```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import datetime
```

```python
url = 'https://www.flipkart.com/search?q=phones%20under%2010000'
```

```python
res = requests.get(url).content
```

```python
soup = BeautifulSoup(res,'lxml')
```

```python
# print(soup.prettify())
```

```python
phoneNames = []
stars = []
reviewRatings = []
offerPrices = []
originalPrices = []
discounts = []
features = []
```

```python
for phone ,star ,reviewRating, offerPrice,feature,originalPrice, discount in zip(soup.find_all('div',class_='_3wU53n'),soup.find_all('div',class_='hGSR34'),soup.find_all('span',class_='_38sUEc'),soup.find_all('div',class_='_1vC4OE _2rQ-NK'),soup.find_all('div',class_='_3ULzGw'),soup.find_all('div',class_='_3auQ3N _2GcJzG'),soup.find_all('div',class_='VGWI6T')): 
    phoneNames.append(phone.text)
    stars.append(star.text)    
    reviewRatings.append(reviewRating.span.span.text)    
    offerPrices.append(offerPrice.text)
    features.append(feature.text)
    originalPrices.append(originalPrice.text)
    discounts.append(discount.span.text)
    
```

```python
data = {
    'Name':phoneNames,
    'star rating':stars,
    'Ratings':reviewRatings,
    'Offer Price':offerPrices,
    'Original Price':originalPrices,
    'Discount':discounts,
    'Specifications':features
}
```

```python
df = pd.DataFrame(data)
```
