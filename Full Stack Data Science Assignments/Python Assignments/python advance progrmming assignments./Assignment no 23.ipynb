```python
class Sudoku:
    def __init__(self, board_string):
        self.board = [[int(board_string[i * 9 + j]) for j in range(9)] for i in range(9)]

    def get_row(self, n):
        return self.board[n]

    def get_col(self, n):
        return [self.board[i][n] for i in range(9)]

    def get_sqr(self, *args):
        if len(args) == 1:
            n = args[0]
            row, col = (n // 3) * 3, (n % 3) * 3
        else:
            row, col = (args[0] // 3) * 3, (args[1] // 3) * 3
        return [self.board[i][j] for i in range(row, row + 3) for j in range(col, col + 3)]
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    def add_data(self, values):
        current = self
        for val in values:
            current.next = ListNode(val)
            current = current.next

    def get_data(self):
        data, current = [], self
        while current:
            data.append(current.val)
            current = current.next
        return data

def add_two_numbers(l1, l2):
    dummy = ListNode()
    current, carry = dummy, 0

    while l1 or l2 or carry:
        val1, val2 = (l1.val if l1 else 0), (l2.val if l2 else 0)
        carry, sum_val = divmod(val1 + val2 + carry, 10)
        current.next = ListNode(sum_val)
        current = current.next
        l1, l2 = (l1.next if l1 else None), (l2.next if l2 else None)

    return dummy.next
```

```python
class CoffeeShop:
    def __init__(self, name, menu):
        self.name = name
        self.menu = menu
        self.orders = []

    def add_order(self, item_name):
        if any(item["name"] == item_name for item in self.menu):
            self.orders.append(item_name)
            return "Order added!"
        return "This item is currently unavailable!"

    def fulfill_order(self):
        if self.orders:
            return f"The {self.orders.pop(0)} is ready!"
        return "All orders have been fulfilled!"

    def list_orders(self):
        return self.orders

    def due_amount(self):
        return sum(item["price"] for item in self.menu if item["name"] in self.orders)

    def cheapest_item(self):
        return min(self.menu, key=lambda x: x["price"])["name"]

    def drinks_only(self):
        return [item["name"] for item in self.menu if item["type"] == "drink"]

    def food_only(self):
        return [item["name"] for item in self.menu if item["type"] == "food"]
```

```python
import sympy

def loneliest_number(lo, hi):
    primes = list(sympy.primerange(lo - 10, hi + 10))
    max_distance, lonely_number, closest_prime = -1, None, None

    for num in range(lo, hi + 1):
        distances = {abs(num - p): p for p in primes}
        min_dist = min(distances)
        closest_p = distances[min_dist]
        if min_dist > max_distance:
            max_distance, lonely_number, closest_prime = min_dist, num, closest_p

    return {"number": lonely_number, "distance": max_distance, "closest": closest_prime}
```

```python
import pickle

class Selfie:
    def __init__(self):
        self._history = []

    def save_state(self):
        self._history.append(pickle.dumps(self.__dict__))

    def recover_state(self, n):
        if 0 <= n < len(self._history):
            new_selfie = Selfie()
            new_selfie.__dict__ = pickle.loads(self._history[n])
            new_selfie._history = self._history.copy()
            return new_selfie
        return self

    def n_states(self):
        return len(self._history)
```
